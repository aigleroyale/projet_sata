SPARK SQL

```python
# Filtrer en excluant la couleur E

query = """
select *
from diamonds
where color != 'E';
"""

resultat = spark.sql(query)
resultat.show()
```

```python
# Calculer le prix minimum, le prix maximum, le prix moyen en arrondissant à 2 après la virgule pour l'ensemble des diamants

query = """
select round(min(price), 2) as minPrice, round(max(price), 2) as maxPrice, round(avg(price), 2) as avgPrice
from diamonds
"""

resultat = spark.sql(query)
resultat.show()
```

```python
# Calculer le prix minimum, le prix maximum, le prix moyen en arrondissant à 2 après la virgule par couleur

query = """
select color, round(min(price), 3) as minPrice, round(max(price), 3) as maxPrice, round(avg(price), 3) as avgPrice
from diamonds
group by color;
"""

resultat = spark.sql(query)
resultat.show()
```

```python
# Enregistrer son dernier résultat (Exercice 1.3)

resultat.write.csv("res.csv")
```
