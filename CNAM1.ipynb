```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
np.random.seed(0)

X = 2*np.random.rand(100, 1)

y = np.sin(X[:,0]) + 0.15 * np.random.randn(100)
```

```python
plt.figure(figsize = (6, 4))
plt.scatter(X, y, s = 50)
plt.xlabel("$X$")
plt.ylabel("$y$")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=0)

plt.figure(figsize = (6,4))
plt.scatter(x_train, y_train, s = 50, edgecolor = 'none', label = 'Entrainement')
plt.scatter(x_test, y_test, c = 'none', s = 50, edgecolor = 'blue', label = 'Evaluation')
plt.xlabel("$X$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
# Regression linéaire

from sklearn import linear_model

reg = linear_model.LinearRegression()

reg.fit(x_train, y_train)

coeff_train = reg.score(x_train, y_train)

print(f"Coefficient de détermination R^2 en train : {coeff_train:2f}")
```

```python
coeff_test = reg.score(x_test, y_test)

print(f"Coefficient de détermination R^2 en test : {coeff_test:2f}")
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x_train,y_train, s=50, edgecolors='none', label="Exemples d'apprentissage")
plt.scatter(x_test,y_test, c='none', s=50, edgecolors='blue', label="Exemples d'évaluation")
plt.xlabel("$x$")
plt.ylabel("$y$")

x_min, x_max = plt.xlim()
nx = 100
xx = np.linspace(x_min, x_max, nx).reshape(-1, 1)
plt.plot(xx, reg.predict(xx), color = 'red', label = 'Regression linéaire')
plt.title("Regression linéaire")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_pred = reg.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

print(f"{mse : .5}")
```
