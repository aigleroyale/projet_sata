```python
from pyspark import SparkFiles
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator

spark = SparkSession.builder \
    .appName("Linear Regression with PySpark MLlib") \
    .getOrCreate()
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"

spark.sparkContext.addFile(url)
```

```python
data = spark.read.csv(SparkFiles.get("BostonHousing.csv"), header=True, inferSchema=True)

data.show()
```

```python
assembler = VectorAssembler(
    inputCols = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "b", "lstat"],
    outputCol = "features"
)
```

```python
data = assembler.transform(data)
```

```python
data.show()
```

```python
final_data = data.select("medv", "features")

final_data.take(2)
```

```python
train_data, test_data = final_data.randomSplit([0.8, 0.2], seed = 123)
```

```python
train_data.take(2)
```

```python
lr = LinearRegression(featuresCol="features", labelCol="medv", predictionCol="predict_medv")
lr_model = lr.fit(train_data)
```

```python
predictions = lr_model.transform(test_data)
```

```python
predictions.take(2)
```

```python
evaluator = RegressionEvaluator(labelCol="medv", predictionCol="predict_medv", metricName="rmse")
```

```python
rmse = evaluator.evaluate(predictions)
```

```python
rmse
```

```python
evaluator2 = RegressionEvaluator(labelCol="medv", predictionCol="predict_medv", metricName="r2")
```

```python
r2 = evaluator2.evaluate(predictions)
```

```python
print(r2)
```

```python
coefficients = lr_model.coefficients
print(coefficients)
```

```python
print(lr_model.intercept)
```

```python
feature_importance = sorted(list(zip(data.columns[:-1], map(abs, coefficients))), key=lambda x: x[1], reverse=True)
```

```python
for feature, importance in feature_importance:
  print(feature, importance)
```

```python
lr_model.save("lr_model")
```

```python
from pyspark.sql import SparkSession
from pyspark import SparkFiles
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.feature import VectorAssembler
```

```python
spark = SparkSession.builder \
    .appName("LogisticRegression with PySpark MLlib") \
    .getOrCreate()
```

```python
url = "https://raw.githubusercontent.com/pkmklong/Breast-Cancer-Wisconsin-Diagnostic-DataSet/master/data.csv"

spark.sparkContext.addFile(url)
```

```python
df = spark.read.csv(SparkFiles.get("data.csv"), header=True, inferSchema=True)

df.show(2)
```

```python
len(df.columns)
```

```python
columns = ['id', 'diagnosis'] + [f'feature_{i}' for i in range(1, len(df.columns) -1)]
```

```python
columns
```

```python
data = df.toDF(*columns)
```

```python
data = data.withColumn("label", (data["diagnosis"] == "M").cast("integer")).drop("diagnosis")

data.show(2)
```

```python
feature_columns = [f"feature_{i}" for i in range(1, 25)]
```

```python
feature_columns
```

```python
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
```

```python
data = assembler.transform(data)
```

```python
data.show(2)
```

```python
train_data, test_data = data.randomSplit([0.8, 0.2], seed = 42)
```

```python
logistic_regression = LogisticRegression(featuresCol="features", labelCol="label")
```

```python
model = logistic_regression.fit(train_data)
```

```python
print(model.coefficients)
```

```python
print(model.intercept)
```

```python
predictions = model.transform(test_data)
```

```python
evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="label")
auc = evaluator.evaluate(predictions)
```

```python
auc
```

```python
# Accuracy, Precision, Recall

multi_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

accuracy = multi_evaluator.evaluate(predictions, {multi_evaluator.metricName : "accuracy"})
```

```python
accuracy
```

```python
precision = multi_evaluator.evaluate(predictions, {multi_evaluator.metricName : "weightedPrecision"})
```

```python
print(precision)
```

```python
recall = multi_evaluator.evaluate(predictions, {multi_evaluator.metricName : "weightedRecall"})
```

```python
print(recall)
```
